```python
# Re-running the code to generate and display the graph again

import matplotlib.pyplot as plt
import numpy as np

# Moment distribution for 3-point bending
x_3point = np.linspace(0, 10, 100)
moment_3point = np.piecewise(x_3point, [x_3point < 5, x_3point >= 5], [lambda x: x, lambda x: 10 - x])

# Moment distribution for pure bending
x_pure = np.linspace(0, 10, 100)
moment_pure = np.full_like(x_pure, 5)

# Plotting the moment distributions
plt.figure(figsize=(12, 6))

# 3-point bending
plt.subplot(1, 2, 1)
plt.plot(x_3point, moment_3point, label='Moment Distribution')
plt.title('3-Point Bending')
plt.xlabel('Length along the beam')
plt.ylabel('Moment')
plt.grid(True)
plt.legend()

# Pure bending
plt.subplot(1, 2, 2)
plt.plot(x_pure, moment_pure, label='Moment Distribution', color='orange')
plt.title('Pure Bending')
plt.xlabel('Length along the beam')
plt.ylabel('Moment')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Ideal moment distribution for 3-point bending
x_3point_ideal = np.linspace(0, 10, 100)
moment_3point_ideal = np.piecewise(x_3point_ideal, [x_3point_ideal < 5, x_3point_ideal >= 5], [lambda x: x, lambda x: 10 - x])

# Practical moment distribution for 3-point bending
moment_3point_practical = np.piecewise(x_3point_ideal, [x_3point_ideal < 5, x_3point_ideal >= 5],
                                      [lambda x: x + np.random.normal(0, 0.5, len(x_3point_ideal)//2),
                                       lambda x: 10 - x + np.random.normal(0, 0.5, len(x_3point_ideal)//2)])

# Ideal moment distribution for pure bending
x_pure_ideal = np.linspace(0, 10, 100)
moment_pure_ideal = np.full_like(x_pure_ideal, 5)

# Practical moment distribution for pure bending
moment_pure_practical = 5 + np.random.normal(0, 0.2, len(x_pure_ideal))

# Plotting the moment distributions
plt.figure(figsize=(12, 10))

# Ideal 3-point bending
plt.subplot(2, 2, 1)
plt.plot(x_3point_ideal, moment_3point_ideal, label='Ideal Moment Distribution')
plt.title('Ideal 3-Point Bending')
plt.xlabel('Length along the beam')
plt.ylabel('Moment')
plt.grid(True)
plt.legend()

# Practical 3-point bending
plt.subplot(2, 2, 2)
plt.plot(x_3point_ideal, moment_3point_practical, label='Practical Moment Distribution', color='red')
plt.title('Practical 3-Point Bending')
plt.xlabel('Length along the beam')
plt.ylabel('Moment')
plt.grid(True)
plt.legend()

# Ideal pure bending
plt.subplot(2, 2, 3)
plt.plot(x_pure_ideal, moment_pure_ideal, label='Ideal Moment Distribution', color='orange')
plt.title('Ideal Pure Bending')
plt.xlabel('Length along the beam')
plt.ylabel('Moment')
plt.grid(True)
plt.legend()

# Practical pure bending
plt.subplot(2, 2, 4)
plt.plot(x_pure_ideal, moment_pure_practical, label='Practical Moment Distribution', color='green')
plt.title('Practical Pure Bending')
plt.xlabel('Length along the beam')
plt.ylabel('Moment')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
